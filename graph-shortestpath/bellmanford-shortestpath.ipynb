For all edges, if the distance to the 
destination can be shortened by 
taking the edge, the distance is 
updated to the new lower value

V-1 times  we make relaxation

Since the longest possible path without a 
cycle can be V-1 edges, the edges must
be scanned V-1  times to ensure the 
shortest path has been found for all nodes

A final scan of all the edges is performed 
and if any distance is updated  means
there is a negative cycle !!!

Applications:
    
Cycle detection can prove to be very important

Negative cycles as well  we have to run the Bellman-Ford algorithm that can handle negative edge weights by default

On the FOREX market it can detect arbitrage situations !!!
